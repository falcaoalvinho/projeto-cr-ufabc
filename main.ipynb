```python
import requests
import networkx as nx
import matplotlib.pyplot as plt

url = input('Insira a URL do repositório: ')

response = requests.get(url)
data = response.json()

grafo_teste = nx.MultiDiGraph()

if response.status_code == 200:
    for transferencias in data['transferências']:
        # if grafo_teste.number_of_nodes() < 200:
        #     grafo_teste.add_edge(transferencias['de'], transferencias['para'])
        # else: 
        #     break

        grafo_teste.add_edge(transferencias['de'], transferencias['para'])

pos = nx.spring_layout(grafo_teste,k=0.5, scale=30)

print(f'Vértices do Grafo: {grafo_teste.number_of_nodes()}')
print(f'Arestas do Grafo: {grafo_teste.number_of_edges()}')
nx.draw(grafo_teste, pos, node_size=30, edge_color='gray', arrowsize=10)
# plt.figure(figsize=(300,300))
plt.show
```
